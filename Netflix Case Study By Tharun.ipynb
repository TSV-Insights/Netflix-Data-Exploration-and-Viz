```python
#Calculating Run times of TV Shows:
tvshow_duration = netflix_df.loc[netflix_df['type'] == 'TV Show','duration'].value_counts().reset_index()

# Binning the seasons
tvshow_duration.replace({'1 Season':'1S',
 '2 Seasons':'2S to 3S','3 Seasons':'2S to 3S',
 '4 Seasons':'4S to 6S','5 Seasons':'4S to 6S','6 Seasons':'4S to 6S',
 '7 Seasons':'7S to 9S','8 Seasons':'7S to 9S','9 Seasons':'7S to 9S',
 '10 Seasons':'10S to 14S','11 Seasons':'10S to 14S','12 Seasons':'10S to 14S','13 Seasons':'10S to 14S',
 '14 Seasons':'15S to 17S','15 Seasons':'15S to 17S','16 Seasons':'15S to 17S','17 Seasons':'15S to 17S'},
 inplace = True)

#Grouping
tvshow_duration=tvshow_duration.groupby('duration')['count'].sum().sort_values(ascending=False).reset_index()
tvshow_duration.rename(columns={'duration':'Seasons'},inplace=True)
tvshow_duration
```

3. 4 to 6 Seasons (193 shows): Typically suggests shows that are moderately successful and have developed a strong following.

4. 7 to 9 Seasons (49 shows): Indicating that long-running shows (lasting 7-9 seasons) are relatively rare on Netflix.

5. 10 to 14 Seasons (14 shows): Suggesting that the platform focuses more on shorter-to-medium-length content, rather than long-lasting series.

6. 15 to 17 Seasons (3 shows): Only a handful of shows make it to this length, emphasizing the rarity of extremely long-running shows on the platform.

**Insights:**

1. Given the popularity of limited series and one-season shows, Netflix should continue to release miniseries and self-contained stories across a variety of genres.

2. As the 2-3 season shows represent a significant portion of the library, Netflix could focus on content that is not too short but also doesn’t stretch on for too long.Continue investing in shows that fall within the 2-3 seasons range, as these have the potential to develop strong followings without risking long-term viewer fatigue.

3. Netflix can continue analyzing the performance of long-running series. If a show is performing well but nearing its end (after 3-4 seasons), Netflix could consider whether extending the show for another season or two is viable based on viewer demand.

4. Netflix can diversify its offerings and maintain a balance between quick-viewing content and more immersive, long-form storytelling.

```python
movie_duration = netflix_df.loc[netflix_df['type'] == 'Movie','duration'].value_counts().reset_index()
#removing 'min' from every entry to help in binning process
movie_duration['duration'] = movie_duration['duration'].str[:-3]
#converting the movie duration entries into integers
movie_duration['duration'] = movie_duration['duration'].astype('int')
#binning the values into categories
bin_range = [0,30,90,150,210,float('inf')]
bin_labels = ['Short Films', 'Medium Films', 'Standard Films', 'Long Films ', 'Epic Films ']
movie_duration['duration'] = pd.cut(movie_duration['duration'],bins = bin_range,labels = bin_labels)
#grouping the different movie types and finding the count
movie_duration = movie_duration.groupby('duration')['count'].sum().sort_values(ascending = False).reset_index()
movie_duration.rename({'duration':'Film Length'},axis = 1,inplace = True)
movie_duration
```

**Insights:**

1. Standard Films make up the largest portion of Netflix's film catalog, Netflix should continue to prioritize this format. Films that are approximately 90-120 minutes remain the most commercially viable

2. Medium Films (120 to 150 minutes) are the second largest category, Netflix could focus on producing or acquiring more films in this range. These films offer the opportunity for deeper narratives and more developed plots without the commitment of a full 2-3 hour epic. Genres like dramas, comedies, and thrillers could benefit from this format.

3. Films like historical epics, fantasy adventures, and long-form documentaries could be more heavily promoted to audiences looking for longer narratives.

4. Netflix could look into promoting short-form content, including experimental works, animation, or festival selections. This could be a key area to explore as part of a short-form content strategy to diversify offerings on the platform, particularly for younger or more experimental viewers.

5. A series of multi-episode films (like a “Lord of the Rings” style trilogy) could work well for the platform and draw in audiences looking for immersive cinematic experiences.

6. Netflix could consider investing in or acquiring big-budget, multi-hour releases in genres like historical epics, fantasy, and adventure. These films can serve as high-profile releases and cultural events.

## **Content Split Analysis**


*   Content split for Top 10 Countries which have produced the most Movies and most TV Shows on Netflix.




```python
# Content Split--- Countries with both TV Shows and Movies
content=netflix_unnested_df.groupby('country')['title'].nunique().sort_values(ascending = False).reset_index().loc[0:10]
content=content.drop(2)  # Dropping Unknown Row

content_merge = pd.merge(content,netflix_movie, on = 'country', how = 'left') # Joining Total counts ,movie counts and TV Show counts with respect to countries
content_merge = pd.merge(content_merge,netflix_tvshow, on = 'country',how = 'left')

content_merge.rename(columns={'title_x':'Total_Count','title_y':'Movie_Count','title':'TVShow_Count'},inplace=True) # Renaming the columns

content_merge['Movie_Count'].fillna(content_merge['Total_Count']-content_merge['TVShow_Count'],inplace = True)# Filling null values
content_merge['TVShow_Count'].fillna(content_merge['Total_Count']-content_merge['Movie_Count'],inplace = True)

content_merge['Movie%'] = round((content_merge['Movie_Count']/content_merge['Total_Count'])*100) # Calculating Split %
content_merge['TV%'] = round((content_merge['TVShow_Count']/content_merge['Total_Count'])*100)

# Count acquired is in Float datatype, changing into Integer type
for i in content_merge.columns[1:]:
 content_merge[i] = content_merge[i].astype('int')

content_merge=content_merge.sort_values(by='Movie%') # Sorting based on Movie% Column
content_merge
```

```python
#setting the plot style
fig,ax = plt.subplots(figsize = (15,8))

#plotting the visual
ax.barh(content_merge['country'],width = content_merge['Movie%'],color = 'orange')
ax.barh(content_merge['country'],width = content_merge['TV%'],left = content_merge['Movie%'],color ='skyblue')
ax.set(xlim=(0,100))

#customizing ticks
ax.set_xticks([])
ax.set_yticklabels(content_merge['country'],fontweight = 'bold',fontfamily='serif')

#adding % values in the bars
for i in content_merge.index:
 ax.text((content_merge.loc[i,'Movie%'])/2,content_merge.loc[i,'country'],f"{content_merge.loc[i,'Movie%']}%",va = 'center', ha='center',fontsize=15, fontweight='light', fontfamily='serif',color='white')
 ax.text((content_merge.loc[i,'Movie%'] + (content_merge.loc[i,'TV%']/2)),content_merge.loc[i,'country'],f"{content_merge.loc[i,'TV%']}%",va = 'center', ha='center',fontsize=15, fontweight='light', fontfamily='serif',color='white')

ax.set_title('Top 10 Countries Movie & TV Show split',
 {'font':'serif', 'size':15,'weight':'bold'})

ax.legend(['Movie','TV Show'],loc = (0.75,1),ncol = 2,fontsize = 13)
plt.show()
```

**Insights:**

1. TV Show Dominance: Countries like South Korea, Japan, and South Korea focus more on TV shows than movies. Netflix should continue to leverage K-drama, anime, and Asian TV series.
2. High-Volume Movie Markets: The United States, India, France, and Germany are leading in movie production. Netflix should continue to expand international movie content, focusing on diverse genres from these countries.
3. Spanish Content Growth: Spain and Mexico are strong in movies, but Spain has started gaining traction in TV shows. Spanish-language content (films and series) should remain a priority.
4. Balancing Genres: In Canada and United Kingdom, where both movies and TV shows are strong, Netflix can invest in a balanced content strategy by pushing both English-language films and original British series.
5. French and German Films: Both countries produce a larger share of movies, so Netflix should ensure a strong presence in French-language and German-language films.

# **Target Audience**

**How Netflix Appeals to Different Audiences: Movies & TV Shows**

```python
#Target Audience Distribution
def age_dist(rating):
  if rating in ['TV-Y', 'TV-G', 'G']:
    return 'Kids'
  elif rating in ['TV-Y7', 'TV-Y7-FV', 'TV-PG', 'PG']:
    return 'Schooled Aged Children'
  elif rating in ['PG-13','TV-14']:
    return 'Teens'
  else:
    return 'Adults'

#For Exploring Movies and TV shows
movie_rating = netflix_df.loc[netflix_df['type'] == 'Movie','rating'].value_counts().reset_index()
tv_rating = netflix_df.loc[netflix_df['type'] == 'TV Show','rating'].value_counts().reset_index()
movie_rating['age'] = movie_rating['rating'].apply(lambda x:age_dist(x))
age_bin_movies = movie_rating.groupby('age')['count'].sum().sort_values(ascending = False).reset_index()
tv_rating['age'] = tv_rating['rating'].apply(lambda x:age_dist(x))
age_bin_tvshows = tv_rating.groupby('age')['count'].sum().sort_values(ascending = False).reset_index()
```

```python
age_bin_movies
```

```python
age_bin_tvshows
```

```python
fig = plt.figure(figsize = (12,6))
mt = fig.add_gridspec(1,2)
axes_0 = fig.add_subplot(mt[0,0])
axes_0.pie(age_bin_movies['count'],labels = age_bin_movies['age'],autopct = '%.1f%%',shadow = True,explode = [0.1,0,0,0])
axes_0.set_title('Movies Vs Audience Age Group')

axes_1 = fig.add_subplot(mt[0,1])
axes_1.pie(age_bin_tvshows['count'],labels = age_bin_tvshows['age'],autopct = '%.1f%%',shadow = True,explode = [0.1,0,0,0])
axes_1.set_title('TV Shows Vs Audience Age Group')

fig.suptitle('Netflix Target Audience Analysis')
plt.show()
```

Insights:

1.Movies:Netflix has 48% of its content for adults audience, 31% for teenagers and rest for children.

2.TV Show:Same as movies, around 43% of its content is for adult audience, but comaratively tv shows hold more place for kids.
